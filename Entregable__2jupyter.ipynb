```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv('match_report_2021-22.csv')

# Mostrar las primeras filas para ver la estructura
print(df.head())
```

```python
print("Forma del dataset:", df.shape)
```

```python
# Filtrar partidos con resultado de empate (D)
df_empates = df[df['Result'] == 'D']

# Mostrar las primeras filas del nuevo dataset
print("\nPartidos con resultado de empate:")
print(df_empates.head())
```

```python
# Filtrar partidos perdidos y como visitante
df_perdidos_visitante = df[(df['Result'] == 'L') & (df['Venue'] == 'Away')]

# Mostrar resultados
print("\nPartidos perdidos como visitante:")
print(df_perdidos_visitante.head())

# Guardar en un nuevo archivo
df_perdidos_visitante.to_csv('partidos_perdidos_visitante.csv', index=False)

print("\nArchivo CSV guardado: partidos_perdidos_visitante.csv")
```

```python
# Ordenar los partidos: los ganados ('W') primero
# Creamos una categoría personalizada para ordenar la columna 'Result'
orden_personalizado = {'W': 0, 'D': 1, 'L': 2}  # W (primero), D (segundo), L (tercero)
df['Result'] = df['Result'].astype('category')  # Convertimos a tipo categórico
df['Result'] = df['Result'].cat.set_categories(orden_personalizado)  # Asignamos el orden

# Aplicamos el ordenamiento
df_ordenado = df.sort_values('Result')

#  Mostrar los primeras filas del nuevo dataset ordenado
print("\nPrimeros partidos después del ordenamiento:")
print(df_ordenado.head())
```

```python
# Guardar el dataset ordenado en un archivo CSV
df_ordenado.to_csv('partidos_ordenados_por_resultado.csv', index=False)

print("\nNuevo dataset guardado: partidos_ordenados_por_resultado.csv")
```

```python
import pandas as pd
from tabulate import tabulate

# Cargar el dataset
df = pd.read_csv('match_report_2021-22.csv')

# Definir orden personalizado para Result: W -> D -> L
orden_personalizado = {'W': 0, 'D': 1, 'L': 2}
df['Result'] = df['Result'].astype('category').cat.set_categories(orden_personalizado)

# Ordenar por resultado (W primero)
df_ordenado = df.sort_values('Result')

# Mostrar tabla con formato usando tabulate
print("\nDatos con tabulate:")
print(tabulate(df_ordenado.head(50), headers='keys', tablefmt='pretty'))
```

```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv('match_report_2021-22.csv')

# Verificar columnas disponibles
print("Columnas disponibles:", df.columns.tolist())

# Reemplazar valores nulos en GF y GA con 0
df['GF'] = df['GF'].fillna(0)
df['GA'] = df['GA'].fillna(0)

# Calcular la diferencia de goles
df['Goal Difference'] = df['GF'] - df['GA']

# Filtrar partidos perdidos por 2 o más goles
df_derrotas_mayor_2 = df[(df["Result"] == "L") & (df["Goal Difference"] <= -2)]

# Mostrar resultados
if not df_derrotas_mayor_2.empty:
    print("\nPartidos perdidos por 2 o más goles:")
    print(df_derrotas_mayor_2[["Date", "Opponent", "GF", "GA", "Goal Difference", "Venue", "Result"]])
else:
    print("No se encontraron partidos que cumplan la condición.")

# Guardar el nuevo dataset
df_derrotas_mayor_2.to_csv('partidos_perdidos_por_2_o_mas_goles.csv', index=False)
```
